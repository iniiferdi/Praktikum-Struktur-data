> **ðŸ’¡ Excercies 1**  
> *Buatlah program yang menampilkan jadwal mata kuliah Anda di hari Jumat dengan jumlah mata kuliah sesuai digit terakhir NPM Anda!*

---

**âœ¨ Solution**

Python Implementation:

```python
class CourseNode:
    def __init__(self, course_name):
        self.course_name = course_name
        self.next = None

class CourseList:
    def __init__(self):
        self.head = None

    def add_course(self, course_name):
        new_course = CourseNode(course_name)
        if not self.head:
            self.head = new_course
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_course

    def show_courses(self):
        if not self.head:
            print("No Course Schedule")
            return
        
        print("List of Courses on Friday:")
        current = self.head
        count = 1
        
        while current:
            print(f"{count}. {current.course_name}")
            current = current.next
            count += 1

course_list = CourseList()

last_digit_npm = int(input(f"Enter amount course"))

for i in range(last_digit_npm):
    data_courses = input(f"Enter Friday's courses")
    course_list.add_course(data_courses)

course_list.show_courses()
```

> **ðŸ’¡ Excercies 2**  
> *Buatlah program yang menampilkan riwayat navigasi folder dalam File Explorer menggunakan Doubly Linked List. Jumlah folder yang Anda navigasikan sesuai dengan digit terakhir NPM Anda!*

---

**âœ¨ Solution**

Python Implementation:

```python
class PathNode:
    def __init__(self, path):
      self.path = path
      self.prev = None
      self.next = None
      
class PathHistory:
    def __init__(self):
      self.current = None
      
    def visit_path(self, path):
     new_path = PathNode(path)
     if not self.current:
         self.current = new_path
         
     else:
         new_path.prev = self.current
         self.current.next = new_path
         self.current = new_path
         
    def show_history(self):
        if not self:
            print('Not History')
            return
        
        print('History File Exploler')
        temp = self.current
        while temp.prev:
            temp = temp.prev
            
        count = 1   
        while temp:
            print(f"{count}. {temp.path}")
            temp = temp.next
            count += 1
            
    def go_back(self):
     if self.current and self.current.prev:
         self.current = self.current.prev
         print(f"\nYou go Back to{self.current.path}")
     else:
         print('In first path')
         
    def go_forward(self):
     if self.current and self.current.next:
         self.current = self.current.next
         print(f"\nYou forward to{self.current.path}")
     else:
         print('In ending of the path')
         
history = PathHistory()

num_pages = int(input("\nEnter the number of pages visited:"))

for i in range(1, num_pages + 1):
  url = input(f"Enter the path to-{i}:")
  history.visit_path(url)

history.show_history()

while True:
    command = input("\nPress 'b' to go back, Press 'f' to go forward, Press 'x' to exit.\nEnter the command:").lower()

    if command == 'b':
      history.go_back()
    elif command == 'f':
      history.go_forward()
    elif command == 'x':
      print('\nExit File Explorer.')
      break
    else:
      print("\nUnknown command.")
```

> **ðŸ’¡ Excercies 3**  
> *Basic Opertation in Linked List*

---

**âœ¨ Solution**

Python Implementation:

```python
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong
        
    
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node
        
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next
        
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None
        
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None
        
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next
        
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"Elemen {key} tidak ditemukan dalam daftar.")
        
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev
        
    def display(self):
        if not self.head:
            print("Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
        
        
ll = LinkedList()

num_elements = int(input("Masukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

while True:
    print("\nPilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi: "))
        data = int(input("Masukkan data: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
```
