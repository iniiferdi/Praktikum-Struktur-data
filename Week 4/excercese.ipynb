> **💡 Excercies 1**  
> *Buatlah sistem antrian pemesanan tiket bioskop menggunakan Queue berbasis List, di mana setiap pelanggan yang datang akan mendapatkan nomor antrian dan dilayani berdasarkan urutan kedatangan.*

---

**✨ Solution**

Python Implementation:

```python
from tabulate import tabulate

class TicketBioskop:
    def __init__(self, max_capacity=10):
        self.queue = []
        self.max_capacity = max_capacity

    def enqueue(self, ticket_id, grade):
        if grade.lower() == "vip":
            print(f"\n🎟️ Pelanggan VIP {ticket_id} langsung mendapatkan kursi tanpa antrian! 🎉\n")
            return
        
        if len(self.queue) >= self.max_capacity:
            print("\n⚠️ Antrian penuh! Mohon tunggu hingga ada slot kosong.\n")
            return

        self.queue.append(ticket_id)
        print(f"\n✅ Pelanggan {ticket_id} (Reguler) berhasil ditambahkan ke antrian.\n")

    def dequeue(self):
        if not self.is_empty():
            served_customer = self.queue.pop(0)
            print(f"\n✅ Pelanggan {served_customer} telah dilayani. Selamat menikmati film! 🍿\n")
            return served_customer
        print("\n⚠️ Antrian kosong! Tidak ada pelanggan untuk dilayani.\n")
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def show_queue(self):
        print("\n================= 📢 STATUS ANTRIAN 📢 =================")
        if self.is_empty():
            print("🚫 Antrian kosong! Tidak ada pelanggan dalam antrean.\n")
            return
        
        table_data = [[idx + 1, ticket_id] for idx, ticket_id in enumerate(self.queue)]
        print(tabulate(table_data, headers=["No", "Kode Pelanggan"], tablefmt="grid"))
        print(f"\n👥 Jumlah pelanggan dalam antrian: {self.size()} orang\n")
        print(f"🎟️ Pelanggan pertama  : {self.front()}")
        print(f"🎟️ Pelanggan terakhir : {self.rear()}")
        print("========================================================\n")

# Fungsi untuk membuat kode pelanggan berdasarkan nama, NPM, dan grade
def generate_customer_code(name, npm, grade):
    code = name[:2].upper() + npm[-2:]
    return f"{code} - {grade.upper()}"

queue = TicketBioskop()

while True:
    print("\n🎬 Bioskop Ticketing System 🎟️")
    print("===================================")
    print("1. ➕ Tambah antrian tiket")
    print("2. 📋 Tampilkan antrian")
    print("3. ✅ Layani pelanggan pertama")
    print("4. ❌ Keluar")
    print("===================================")

    choice = input("🔹 Pilih menu: ")
    
    if choice == "1":
        name = input("✏️ Masukkan nama mahasiswa: ")
        npm = input("🔢 Masukkan NPM: ")
        grade = input("⭐ Masukkan grade (VIP/Reguler): ").strip().lower()
        
        if len(name) < 2 or len(npm) < 2:
            print("\n⚠️ Nama atau NPM tidak valid! Harap masukkan dengan benar.\n")
            continue

        if grade not in ["vip", "reguler"]:
            print("\n⚠️ Grade tidak valid! Harap masukkan 'VIP' atau 'Reguler'.\n")
            continue
        
        ticket_id = generate_customer_code(name, npm, grade)
        queue.enqueue(ticket_id, grade)

    elif choice == "2":
        queue.show_queue()

    elif choice == "3":
        queue.dequeue()

    elif choice == "4":
        print("\n🎥 Terima kasih telah menggunakan sistem tiket bioskop! Selamat menonton! 🍿\n")
        break

    else:
        print("\n⚠️ Pilihan tidak valid! Silakan pilih menu yang tersedia.\n")
```

> **💡 Excercies 2**  
> *Buatlah sistem antrian layanan fotokopi di kampus menggunakan Queue berbasis List, di mana mahasiswa yang datang lebih dulu akan lebih dulu mendapatkan layanan.*

---

**✨ Solution**

Python Implementation:

```python
from tabulate import tabulate

class FotocopyQueue:
    def __init__(self):
        self.queue = []
    
    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            print(f"\n🚀 Fast Track! Pesanan {order_id} langsung diproses tanpa antri! 🎉\n")
            return
        
        self.queue.append((order_id, num_pages))
        print(f"\n✅ Pesanan {order_id} dengan {num_pages} halaman berhasil ditambahkan ke antrian.\n")
    
    def dequeue(self):
        if not self.is_empty():
            order_id, num_pages = self.queue.pop(0)
            if num_pages > 100:
                print(f"\n⚠️ Pesanan {order_id} memiliki lebih dari 100 halaman, diproses dalam dua tahap...\n")
                self.queue.append((order_id, num_pages - 100))
                print(f"✅ Tahap pertama selesai untuk {order_id}, sisa {num_pages - 100} halaman akan diproses setelahnya.\n")
            else:
                print(f"\n✅ Pesanan {order_id} telah selesai diproses!\n")
        else:
            print("\n⚠️ Antrian kosong! Tidak ada pesanan untuk diproses.\n")
    
    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"
    
    def is_empty(self):
        return len(self.queue) == 0
    
    def size(self):
        return len(self.queue)
    
    def show_queue(self):
        print("\n================= 📢 STATUS ANTRIAN 📢 =================")
        if self.is_empty():
            print("🚫 Antrian kosong! Tidak ada pesanan dalam antrean.\n")
            return
        
        table_data = [[idx + 1, order_id, pages] for idx, (order_id, pages) in enumerate(self.queue)]
        print(tabulate(table_data, headers=["No", "ID Pemesanan", "Jumlah Halaman"], tablefmt="grid"))
        print(f"\n👥 Jumlah pesanan dalam antrian: {self.size()}\n")
        print(f"📌 Pesanan pertama  : {self.front()[0]}")
        print(f"📌 Pesanan terakhir : {self.rear()[0]}")
        print("========================================================\n")

# Fungsi untuk membuat ID pemesanan
def generate_order_id(name, npm):
    return name[:3].upper() + npm[-2:]

queue = FotocopyQueue()

while True:
    print("\n📄 Sistem Antrian Fotokopi 🖨️")
    print("===================================")
    print("1. ➕ Tambah pesanan fotokopi")
    print("2. 📋 Tampilkan antrian")
    print("3. ✅ Proses pesanan pertama")
    print("4. ❌ Keluar")
    print("===================================")

    choice = input("🔹 Pilih menu: ")
    
    if choice == "1":
        name = input("✏️ Masukkan nama mahasiswa: ")
        npm = input("🔢 Masukkan NPM: ")
        try:
            num_pages = int(input("📄 Masukkan jumlah halaman: "))
        except ValueError:
            print("\n⚠️ Harap masukkan angka untuk jumlah halaman!\n")
            continue
        
        if len(name) < 3 or len(npm) < 2:
            print("\n⚠️ Nama atau NPM tidak valid! Harap masukkan dengan benar.\n")
            continue
        
        order_id = generate_order_id(name, npm)
        queue.enqueue(order_id, num_pages)
    
    elif choice == "2":
        queue.show_queue()
    
    elif choice == "3":
        queue.dequeue()
    
    elif choice == "4":
        print("\n📃 Terima kasih telah menggunakan sistem antrian fotokopi! 🖨️\n")
        break
    
    else:
        print("\n⚠️ Pilihan tidak valid! Silakan pilih menu yang tersedia.\n")
```

> **💡 Excercies 3**  
> *Buatlah sistem antrian kendaraan masuk ke parkiran mall menggunakan Queue berbasis List, di mana kendaraan masuk dan keluar berdasarkan urutan kedatangannya.*

---

**✨ Solution**

Python Implementation:

```python
from tabulate import tabulate

class ParkingQueue:
    def __init__(self, max_capacity=15):
        self.queue = []
        self.max_capacity = max_capacity

    def enqueue(self, vehicle_number, is_electric=False):
        if is_electric:
            self.queue.insert(0, vehicle_number)
            print(f"\n⚡ Kendaraan listrik {vehicle_number} mendapatkan prioritas parkir! 🚗⚡\n")
        else:
            if len(self.queue) >= self.max_capacity:
                print("\n🚗 Parkiran penuh! Harap tunggu hingga ada slot kosong.\n")
                return
            self.queue.append(vehicle_number)
            print(f"\n✅ Kendaraan {vehicle_number} berhasil masuk ke parkiran.\n")

    def dequeue(self):
        if not self.is_empty():
            exited_vehicle = self.queue.pop(0)
            print(f"\n✅ Kendaraan {exited_vehicle} telah keluar dari parkiran.\n")
            return exited_vehicle
        print("\n⚠️ Parkiran kosong! Tidak ada kendaraan untuk dikeluarkan.\n")
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else "Parkiran kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Parkiran kosong"

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def show_queue(self):
        print("\n================= 🅿️ STATUS PARKIRAN 🅿️ =================")
        if self.is_empty():
            print("🚫 Parkiran kosong! Tidak ada kendaraan dalam antrean.\n")
            return
        
        table_data = [[idx + 1, vehicle] for idx, vehicle in enumerate(self.queue)]
        print(tabulate(table_data, headers=["No", "Nomor Plat"], tablefmt="grid"))
        print(f"\n🚗 Jumlah kendaraan dalam parkiran: {self.size()} unit\n")
        print(f"🚘 Kendaraan pertama  : {self.front()}")
        print(f"🚙 Kendaraan terakhir : {self.rear()}")
        print("========================================================\n")

# Fungsi untuk membuat nomor plat kendaraan berdasarkan kota lahir dan NPM
def generate_vehicle_number(city, npm):
    code = city[:2].upper() + npm[-2:]
    return f"{code}"

queue = ParkingQueue()

while True:
    print("\n🅿️ Sistem Manajemen Parkiran 🚗")
    print("===================================")
    print("1. ➕ Tambah kendaraan ke parkiran")
    print("2. 📋 Tampilkan daftar kendaraan")
    print("3. ✅ Keluarkan kendaraan pertama")
    print("4. ❌ Keluar")
    print("===================================")

    choice = input("🔹 Pilih menu: ")
    
    if choice == "1":
        city = input("🏙️ Masukkan kota lahir: ")
        npm = input("🔢 Masukkan NPM: ")
        is_electric = input("⚡ Apakah kendaraan listrik? (ya/tidak): ").strip().lower() == "ya"
        
        if len(city) < 2 or len(npm) < 2:
            print("\n⚠️ Kota lahir atau NPM tidak valid! Harap masukkan dengan benar.\n")
            continue
        
        vehicle_number = generate_vehicle_number(city, npm)
        queue.enqueue(vehicle_number, is_electric)

    elif choice == "2":
        queue.show_queue()

    elif choice == "3":
        queue.dequeue()

    elif choice == "4":
        print("\n🚗 Terima kasih telah menggunakan sistem parkiran! Selamat berkendara! 🏁\n")
        break

    else:
        print("\n⚠️ Pilihan tidak valid! Silakan pilih menu yang tersedia.\n")
```

> **💡 Excercies 4**  
> *Buatlah sistem antrian pengambilan sertifikat di kampus menggunakan Queue berbasis List, di mana mahasiswa yang telah menyelesaikan seminar atau pelatihan dapat mengambil sertifikat berdasarkan urutan kedatangan.*

---

**✨ Solution**

Python Implementation:

```python
from tabulate import tabulate

class CertificateQueue:
    def __init__(self, max_capacity=20):
        self.regular_queue = []
        self.achiever_queue = []
        self.max_capacity = max_capacity

    def enqueue(self, student_id, is_achiever=False):
        if is_achiever:
            print(f"\n🏆 Mahasiswa {student_id} berprestasi! Sertifikat dapat diambil tanpa antrian. 🎖️\n")
            return
        
        if len(self.regular_queue) + len(self.achiever_queue) >= self.max_capacity:
            print("\n⚠️ Antrian penuh! Sertifikat hanya dapat diambil setelah ada yang dilayani.\n")
            return
        
        if is_achiever:
            self.achiever_queue.append(student_id)
            print(f"\n✅ Mahasiswa {student_id} masuk dalam jalur prestasi.\n")
        else:
            self.regular_queue.append(student_id)
            print(f"\n✅ Mahasiswa {student_id} masuk dalam jalur reguler.\n")

    def dequeue(self):
        if self.achiever_queue:
            served_student = self.achiever_queue.pop(0)
        elif self.regular_queue:
            served_student = self.regular_queue.pop(0)
        else:
            print("\n⚠️ Antrian kosong! Tidak ada mahasiswa untuk dilayani.\n")
            return None
        
        print(f"\n✅ Mahasiswa {served_student} telah mengambil sertifikat.\n")
        return served_student

    def front(self):
        if self.achiever_queue:
            return self.achiever_queue[0]
        elif self.regular_queue:
            return self.regular_queue[0]
        return "Antrian kosong"

    def rear(self):
        if self.regular_queue:
            return self.regular_queue[-1]
        elif self.achiever_queue:
            return self.achiever_queue[-1]
        return "Antrian kosong"

    def is_empty(self):
        return len(self.regular_queue) == 0 and len(self.achiever_queue) == 0

    def size(self):
        return len(self.regular_queue) + len(self.achiever_queue)

    def show_queue(self):
        print("\n================= 🎓 STATUS ANTRIAN 🎓 =================")
        if self.is_empty():
            print("🚫 Antrian kosong! Tidak ada mahasiswa dalam antrean.\n")
            return
        
        table_data = [[idx + 1, student] for idx, student in enumerate(self.achiever_queue + self.regular_queue)]
        print(tabulate(table_data, headers=["No", "Nomor Antrian"], tablefmt="grid"))
        print(f"\n👥 Jumlah mahasiswa dalam antrian: {self.size()} orang\n")
        print(f"🎟️ Mahasiswa pertama  : {self.front()}")
        print(f"🎟️ Mahasiswa terakhir : {self.rear()}")
        print("========================================================\n")

# Fungsi untuk membuat nomor antrian berdasarkan program studi dan NPM
def generate_student_id(program, npm):
    code = program[0].upper() + npm[-2:]
    return f"{code}"

queue = CertificateQueue()

while True:
    print("\n🎓 Sistem Antrian Pengambilan Sertifikat 📜")
    print("===================================")
    print("1. ➕ Tambah mahasiswa ke antrian")
    print("2. 📋 Tampilkan daftar antrian")
    print("3. ✅ Layani mahasiswa pertama")
    print("4. ❌ Keluar")
    print("===================================")

    choice = input("🔹 Pilih menu: ")
    
    if choice == "1":
        program = input("📖 Masukkan program studi: ")
        npm = input("🔢 Masukkan NPM: ")
        is_achiever = input("🏆 Apakah pernah mengikuti kompetisi nasional? (ya/tidak): ").strip().lower() == "ya"
        
        if len(program) < 1 or len(npm) < 2:
            print("\n⚠️ Program studi atau NPM tidak valid! Harap masukkan dengan benar.\n")
            continue
        
        student_id = generate_student_id(program, npm)
        queue.enqueue(student_id, is_achiever)

    elif choice == "2":
        queue.show_queue()

    elif choice == "3":
        queue.dequeue()

    elif choice == "4":
        print("\n🎓 Terima kasih telah menggunakan sistem antrian sertifikat! 📜\n")
        break

    else:
        print("\n⚠️ Pilihan tidak valid! Silakan pilih menu yang tersedia.\n")
```
